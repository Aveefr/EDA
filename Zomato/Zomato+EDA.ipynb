```python
df_country = pd.read_excel('ZomatoDataset\Country-Code.xlsx')
df_country.head()
```

```python
df_final = pd.merge(left=df, right=df_country, on='Country Code', how='left')
df_final.head()
```

```python
df.dtypes
```

```python
country_names = df_final.Country.value_counts().index
```

```python
country_values = df_final.Country.value_counts().values
```

```python
plt.pie(x=country_values, labels=country_names)
```

```python
# Pie Chart - Top 3 countries
plt.pie(x=country_values[:3], labels=country_names[:3], autopct='%1.2f')
```

### Observation :
1. Zomato max records are from India
2. after India, USA and UK are the countries most records are from

```python
df_final.columns
```

```python
ratings = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
ratings
```

### Observation:
- 4.5 - 4.9 --> Excellent
- 4.0 - 4.5 --> Very Good
- 3.5 - 3.9 --> Good
- 2.5 - 3.4 --> Average
- 1.8 - 2.4 --> Poor
- 0 --> Not Rated

```python
ratings.head()
```

```python
palette = ['blue', 'red', 'orange', 'yellow', 'green', 'green']
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color', palette=palette)
```

### Observation :
- Not rated is very high
- max rating is between 2.5 and 3.4

```python
ratings
```

```python
sns.countplot(data=ratings, x='Rating color', palette=palette)
```

```python
# Find the countries name that has given 0 rating
df_final[df_final['Aggregate rating'] == 0]['Country'].value_counts().reset_index()
```

### Observation :
- max no of 0 rating is from India

```python
# Find out which currency is used by which country
df_final[['Country', 'Currency']].value_counts().reset_index().drop('count', axis=1)
```

```python
# Find out which countries has online delivery option
df_final[['Country', 'Has Online delivery']].value_counts().reset_index().drop('count', axis=1)
```

### Observation :
- India & UAE only have online delivery option
